```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
```

```python
df=pd.read_csv('sonar.csv')
df.head(7)
```

```python
x=df.iloc[:,0:60].values
y=df.iloc[:,60].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
# from sklearn.preprocessing import StandardScaler
# sc=StandardScaler()
# x_train=sc.fit_transform(x_train)
# x_test=sc.transform(x_test)
```

```python
#importing the keras libraries
import keras
from keras.models import Sequential
from keras.layers import Dense #connection btw the layers is done using dense function
```

```python
#initialising the ANN
classifier=Sequential()
```

```python
#adding the input layer and the first hidden layer
classifier.add(Dense(output_dim=31,init='uniform',activation='relu',input_dim=60))
```

```python
#adding second hidden layer
classifier.add(Dense(output_dim=31,init='uniform',activation='relu'))
```

```python
#adding the output layer
classifier.add(Dense(output_dim=1,init='uniform',activation='sigmoid'))
```

```python
classifier.compile(optimizer='adam',loss='binary_crossentropy', metrics=['accuracy'])
```

```python
#fitting the Ann to the training set
classifier.fit(x_train,y_train,batch_size=10,nb_epoch=150)
```

```python
y_pred=classifier.predict(x_test)
y_pred=(y_pred>0.5)
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
